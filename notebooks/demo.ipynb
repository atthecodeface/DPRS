# **Demo Rust integration**

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from numpy.typing import NDArray
```

```python
from dprs import sim
sim
```

```python
n_x = 60
n_y = 40
n_iterations = 1_000
```

```python
raw_lattice: list = sim.dp(n_x, n_y, n_iterations)
lattice: NDArray = np.array(raw_lattice, dtype=np.bool,).reshape(n_y, n_x).T
```

```python
color_map = ListedColormap([[0.8,0.8,0.8], [0.7,0,0.7]])
plt.imshow(((lattice.T)), cmap=color_map, origin="lower",)
color_bar = plt.colorbar(ticks=[0.25,.75], shrink=0.5*(n_y/n_x)**0.25, aspect=15,)
color_bar.set_ticklabels([0,1])
plt.grid(ls=":")
```
